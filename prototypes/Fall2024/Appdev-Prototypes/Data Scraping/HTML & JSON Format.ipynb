## HTML FILE FORMAT

```python
import requests
from bs4 import BeautifulSoup


url = 'https://www.consumerfinance.gov/data-research/consumer-complaints/'

# Send a request to fetch the HTML content
response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    # Parse the HTML content using BeautifulSoup
    soup = BeautifulSoup(response.content, 'html.parser')

    # Example: Extracting all paragraphs <p> from the HTML
    paragraphs = soup.find_all('p')

    # Save the HTML content to a file
    with open('scraped_content.html', 'w') as file:
        for paragraph in paragraphs:
            file.write(str(paragraph))  # Write each paragraph into the file
            file.write('\n')  # Add a newline for readability

    print("HTML content saved successfully!")
else:
    print(f"Failed to retrieve the HTML page. Status code: {response.status_code}")
```

## JSON FILE FORMAT

```python
import requests
import json

# URL of the JSON file
url = 'https://data.usaid.gov/api/views/dt8c-833c/rows.json?accessType=DOWNLOAD'

# Send a request to fetch the JSON data
response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    data = response.json()

    # Save the JSON data to a file
    with open('dca_loan_transactions.json', 'w') as file:
        json.dump(data, file, indent=4)

    # Print the first few entries to inspect the data
    print(json.dumps(data['data'][:5], indent=4))
else:
    print(f"Failed to retrieve the data. Status code: {response.status_code}")
```

The JSON code fetches data from an API, saves it to a file, and prints a few entries. The HTML code scrapes a webpage, extracts specific HTML elements (like paragraphs), and saves them into a file. Both codes check if the request is successful before processing the data. The JSON code works with structured data from an API, while the HTML code handles content scraped directly from a webpage.

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import json
```

```python
url = 'https://data.usaid.gov/api/views/dt8c-833c/rows.json?accessType=DOWNLOAD'
# sending request
response = requests.get(url)
```

```python
# Check if the request was successful
if response.status_code == 200:
    data = response.json()

    # Save the JSON data to a file for reference
    with open('dca_loan_transactions.json', 'w') as file:
        json.dump(data, file, indent=4)

    # Check the structure of the data
    if 'data' in data:
        # Extract the actual dataset rows
        rows = data['data']

        # Convert the rows into a DataFrame
        df = pd.DataFrame(rows)

        # Display the first few rows of the DataFrame
        print(df.head())

        # Save the DataFrame to a CSV file
        df.to_csv('dca_loan_transactions.csv', index=False)
        print("Data has been saved to 'dca_loan_transactions.csv'")
    else:
        print("The expected 'data' key was not found in the JSON structure.")
else:
    print(f"Failed to retrieve the data. Status code: {response.status_code}")
```
